# LSTM Exercise Classification – Push-up (Correct vs Incorrect)

Load push-up videos from `data/raw/Correct sequence` and `data/raw/Wrong sequence`, train Conv3D model, save to `models/`.

```python
import sys
from pathlib import Path

ROOT = Path.cwd()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from src.data_loader import load_raw_data, ensure_directories
from src.preprocessing import prepare_data
from src.train import train_model
from src.predict import save_model_and_artifacts
```

```python
ensure_directories('.')
X, y, data_dir = load_raw_data('.')
print('X shape:', X.shape)
print('y shape:', y.shape)
print('Labels:', set(y))
```

```python
X_train, X_val, y_train, y_val, encoder, class_names = prepare_data(X, y)
print('Train:', X_train.shape, 'Val:', X_val.shape)
print('Classes:', class_names)
```

```python
model, val_acc = train_model(X_train, y_train, X_val, y_val, epochs=20, batch_size=4)
print(f'Val accuracy: {val_acc:.4f}')
```

```python
save_model_and_artifacts(
    model, encoder, val_acc,
    models_dir='models',
    input_shape=X_train.shape[1:]
)
print('Model saved to models/ (final_model.keras, encoder.pkl, meta.pkl)')
```
