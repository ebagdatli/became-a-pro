# Human Action Recognition – Final Pipeline
Load data, train CNN, save model to `models/`.

```python
import os
import sys
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split

# Run from HumanActionRecognition folder (competition_path)
ROOT = Path.cwd()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from src.data_loader import load_raw_data, ensure_directories
from src.preprocessing import prepare_dataframes, build_datasets
from src.train import build_model
from src.predict import save_model_and_artifacts
```

```python
ensure_directories('.')
train_df, test_df, data_dir = load_raw_data('.')
print('Train shape:', train_df.shape, 'Test shape:', test_df.shape)
train_df, test_df, categories = prepare_dataframes(train_df, test_df, data_dir)
print('Categories:', categories)
```

```python
train_set, val_set = train_test_split(
    train_df, test_size=0.2, stratify=train_df['label'], random_state=42
)
train_dataset, val_dataset = build_datasets(train_set, val_set, batch_size=32, img_size=128)
print('Datasets built.')
```

```python
model = build_model(img_size=128, num_classes=len(categories))
model.summary()
```

```python
# Reduced epochs to finish within timeout; increase when running locally
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=15
)
```

```python
save_model_and_artifacts(model, categories, history, models_dir='models')
print('Model and artifacts saved to models/')
```
